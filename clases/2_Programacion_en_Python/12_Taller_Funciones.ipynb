## **Ejercicio 4**

Un cajero autom√°tico tiene 500.000 pesos disponibles.
Un usuario puede retirar varias veces, pero no puede retirar m√°s de lo que queda en el cajero. Escribe un programa que:

- Pregunte cu√°nto quiere retirar.

- Reste ese valor al saldo.

- Termine cuando el saldo llegue a 0 o el usuario escriba 0.

```python
saldo = 500000

while saldo > 0:
    print(f"\nSaldo disponible: ${saldo:,}")
    retiro = float(input("¬øCu√°nto desea retirar? (0 para salir): "))

    if retiro == 0:
        print("Operaci√≥n finalizada.")
        break
    elif retiro > saldo:
        print("Error: No hay suficiente saldo disponible.")
    else:
        saldo -= retiro
        print(f"Retiro exitoso. Nuevo saldo: ${saldo:,}")

if saldo == 0:
    print("El cajero se ha quedado sin fondos.")
```

## **Ejercicio 5**

Crea una funci√≥n raiz_cuadrada(num) que calcule la ra√≠z cuadrada de un n√∫mero.

Si el n√∫mero es negativo, devuelve: "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."

Si el valor no es num√©rico, devuelve: "Error: el valor debe ser num√©rico."

NOTA: utiliza import math y la funci√≥n math.sqrt(num). No olvides documentar la funci√≥n.

```python
import math

def raiz_cuadrada(num):
    """
    Calcula la ra√≠z cuadrada de un n√∫mero.

    Par√°metros:
    num: N√∫mero al que se le calcular√° la ra√≠z cuadrada

    Retorna:
    float: Ra√≠z cuadrada del n√∫mero
    str: Mensaje de error si el n√∫mero es negativo o no es num√©rico
    """
    try:
        # Verificar si es num√©rico
        num = float(num)

        # Verificar si es negativo
        if num < 0:
            return "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."

        return math.sqrt(num)

    except (ValueError, TypeError):
        return "Error: el valor debe ser num√©rico."
```

```python
print(raiz_cuadrada(16))
```

```python
print(raiz_cuadrada(-9))
```

```python
print(raiz_cuadrada("abc"))
```

Escribe una funci√≥n llamada es_par que reciba un n√∫mero entero como par√°metro y retorne True si el n√∫mero es par y False si es impar.

NOTA: % es el operador m√≥dulo en Python. Lo que hace es dividir un n√∫mero entre otro y devolver el residuo (lo que sobra de la divisi√≥n entera).

Si al dividir un n√∫mero en 2, el residuo es cero, entonces el n√∫mero es par, de lo contrario, es impar.

```python
def es_par(numero):
    """
    Determina si un n√∫mero es par o impar.

    Par√°metros:
    numero (int): N√∫mero entero a evaluar

    Retorna:
    bool: True si es par, False si es impar
    """
    return numero % 2 == 0
```

```python
print(es_par(7))    # Deber√≠a imprimir False
```

## **Ejercicio 7**

Escribe un c√≥digo que permita al usuario crear una contrase√±a y que verifique que esta cumpla todas las siguientes condiciones:

- Tiene al menos 8 caracteres.

- Contiene al menos una letra may√∫scula.

- Contiene al menos una letra min√∫scula.

- Contiene al menos un n√∫mero.

Si cumple todas las condiciones, muestra "Contrase√±a v√°lida", de lo contrario, explica qu√© condici√≥n no se cumple.

```python
# Verificador de contrase√±as
contrasena = input("Ingrese su contrase√±a: ")

# Inicializar verificadores
tiene_longitud = len(contrasena) >= 8
tiene_mayuscula = False
tiene_minuscula = False
tiene_numero = False

# Verificar cada car√°cter
for caracter in contrasena:
    if caracter.isupper():
        tiene_mayuscula = True
    elif caracter.islower():
        tiene_minuscula = True
    elif caracter.isdigit():
        tiene_numero = True

# Verificar todas las condiciones
if tiene_longitud and tiene_mayuscula and tiene_minuscula and tiene_numero:
    print("Contrase√±a v√°lida")
else:
    print("La contrase√±a no cumple con las siguientes condiciones:")
    if not tiene_longitud:
        print("- Debe tener al menos 8 caracteres")
    if not tiene_mayuscula:
        print("- Debe contener al menos una letra may√∫scula")
    if not tiene_minuscula:
        print("- Debe contener al menos una letra min√∫scula")
    if not tiene_numero:
        print("- Debe contener al menos un n√∫mero")
```

## **Ejercicio 8**

Crea una funci√≥n llamada "extraer_dominio" que extraiga el dominio de una p√°gina web a partir de una cadena que contenga un email de la siguiente forma:

usuario@dominio.com
    
Por ejemplo: si se le pasa "lmcastroco@gmail.com", la funci√≥n retornar√°: gmail.com

```python
def extraer_dominio(email):
    """
    Extrae el dominio de un correo electr√≥nico.

    Par√°metros:
    email (str): Correo electr√≥nico en formato usuario@dominio.com

    Retorna:
    str: El dominio del correo
    """
    partes = email.split('@')
    if len(partes) == 2:
        return partes[1]
    else:
        return "Formato de email inv√°lido"
```

```python
extraer_dominio('juanita.perez@hotmail.com')
```

## **Ejercicio 9**

Escribe una funci√≥n llamada contar_desempleo que reciba una cadena y retorne el n√∫mero de veces que aparece la palabra "desempleo" en el texto. Ten en cuenta que la palabra puede estar en may√∫sculas, min√∫sculas o una mezcla de ellas, pero por ahora no te preocupes por casos especiales como cuando la palabra tiene alg√∫n signo de puntuaci√≥n inmediatamente desp√∫es de √©sta.

```python
def contar_desempleo(texto):
    """
    Cuenta cu√°ntas veces aparece la palabra "desempleo" en un texto.

    Par√°metros:
    texto (str): Texto a analizar

    Retorna:
    int: N√∫mero de veces que aparece "desempleo"
    """
    # Convertir todo a min√∫sculas para comparaci√≥n
    texto_minusculas = texto.lower()
    palabra = "desempleo"

    # Contar apariciones
    contador = texto_minusculas.count(palabra)

    return contador
```

```python
contar_desempleo("El gobierno tiene como meta reducir el n√∫mero de personas desocupadas un 10%.")
```

## **Ejercicio 10**

Vas conduciendo un carro muy r√°pido y eres detenido por un agente de tr√°nsito.

Escribe una funci√≥n "multa" que retorne uno de los 3 siguientes resultados:  

- Si la velocidad es 60 Km/h o menos, el resultado es "Sin multa".
- Si la velocidad est√° entre 60 y 80 Km/h (inclusive), el resultado es "Multa peque√±a".
- Si la velocidad es mayor a 80 Km/h, el resultado es "Multa grande".

Adem√°s, considera que si esto ocurre el d√≠a de tu cumplea√±os (codificado como par√°metro booleano que recibe la funci√≥n), tu velocidad podr√≠a ser 5 Km/ mayor en todos los casos.

```python
def multa(velocidad, es_cumpleanos):
    """
    Determina el tipo de multa seg√∫n la velocidad.

    Par√°metros:
    velocidad (float): Velocidad del veh√≠culo en km/h
    es_cumpleanos (bool): True si es el cumplea√±os del conductor

    Retorna:
    str: Tipo de multa
    """
    # Ajustar velocidad si es cumplea√±os (tolerancia de 5 km/h)
    if es_cumpleanos:
        velocidad -= 5

    # Determinar multa
    if velocidad <= 60:
        return "Sin multa"
    elif velocidad <= 80:
        return "Multa peque√±a"
    else:
        return "Multa grande"
```

```python
multa(81, True)
```

```python
multa(81, False)
```

## **Ejercicio 14**

Tienes un diccionario con exportaciones en millones de USD. Escribe un ciclo for que recorra los valores, calcule el promedio de exportaciones y lo imprima como: "El promedio de exportaciones es: XXXX millones de USD"

```python
exportaciones = {"Colombia": 1500, "M√©xico": 2300, "Chile": 1200, "Per√∫": 1800}
```

## **Ejercicio 15**

Un estudiante invierte 1.000 d√≥lares en un fondo que crece un 5% mensual. Escribe un programa que use while para calcular en cu√°ntos meses su inversi√≥n supera los 2.000 d√≥lares.

```python
inversion = 1000
meta = 2000
tasa_mensual = 0.05
meses = 0

while inversion <= meta:
    inversion *= (1 + tasa_mensual)
    meses += 1
    print(f"Mes {meses}: ${inversion:.2f}")

print(f"La inversi√≥n supera los 2000 d√≥lares en {meses} meses.")
```

## **Ejercicio 16**

Una empresa quiere vender al menos 100 unidades de un producto. Cada d√≠a vende un n√∫mero aleatorio de unidades entre 1 y 10 (pista: usa import random y random.randint(1,10)). Escribe un programa con while que muestre:

- El total vendido cada d√≠a.

- En cu√°ntos d√≠as se logr√≥ la meta.

```python
import random

meta = 100
total_vendido = 0
dias = 0

print("Registro de ventas diarias:")
print("-" * 30)

while total_vendido < meta:
    dias += 1
    venta_dia = random.randint(1, 10)
    total_vendido += venta_dia

    print(f"D√≠a {dias}: Vendidas {venta_dia} unidades - Total acumulado: {total_vendido}")

print("-" * 30)
print(f"\nMeta alcanzada en {dias} d√≠as con un total de {total_vendido} unidades vendidas.")
```
